Para printar os pontos dos dias em promoção, sera nescessário criar uma coluna auxiliar no DataFrame que indique as cores.

1-Vamos criar uma função que receba um texto e baseado nelo retorne um valor em texto para a cor.

Definiremos que :

Se há Promoção : 'green'


Se NÃO há Promoção : 'green'

```python
def cor_da_promo(valor:str):
    if valor == "Nao":
        return 'blue'
    else:
        return 'green'
```

Com a função criada, vamos criar uma nova coluna do DataFrame chamada 'Cor_da_Promocao' que ira receber o retorno da função para cada item da coluna 'Promocao' do DataFrame correspondente.

```python
dados['Cor_da_Promocao'] = dados['Promocao'].apply(cor_da_promo)
```

Imprimindo o novo data Frame, temos:

```python
print(dados)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

import pandas as pd #para manipular os datasets
```

O exemplo manda utilizar uma tabela especifica contendo dados de diabetes.

Vamos carregar a que pede o exempo e logo depois tentamos fazer com nossa tabela.

Escolhendo somente um tipo de diabetes:

```python
diabetes_X = diabetes.data[:,np.newaxis,2] #np.newaxis transformar linhas em colunas. No caso ele vai pegar a segunda coluna da tabela inserida
diabetes_data_frame_x = diabetes_data_frame['bmi'] # o equivalente a diabetes_X = diabetes.data[:,np.newaxis,2]
```

Separa elementos de treinamento e elementos para teste.

```python
diabetes_X_train = diabetes_X[:-20]#pega do primeiro valor ate os ultimos 20
diabetes_X_test = diabetes_X[-20:]#pega os 20 ultimos valores
```

Dividir os alvos da variavel 'diabetes'.

```python
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test = diabetes.target[-20:]
#transformando diabetes.target em um dataframe
diabetes_data_frame_y_ = pd.DataFrame(data=diabetes.target[:, np.newaxis])
```

Criar um objeto de regreção linear:

# Segundo Metodo de Regreção Linear Simples Usando o DataFrame do Pandas
Fonte: https://www.linkedin.com/pulse/regress%C3%A3o-linear-simples-e-m%C3%BAltipla-em-python-100-pr%C3%A1tico-santos/?originalSubdomain=pt

```python
dados.plot.scatter(x='Preco_Cafe', y='Vendas_Cafe', title='Preço do Cafe vs Vendas', ylabel='Vendas do Café', xlabel='Preço do Café (R$)')
```

Converte os valores de X e Y para arrays.

```python
x = dados['Preco_Cafe'].values.reshape(-1,1) #converte a coluda do dataframe em array os numeors como argumetnos sao linhas e colunas.
y = dados['Vendas_Cafe'].values.reshape(-1,1)
```

Cria o objeto regressão Linear.

```python
regressao = linear_model.LinearRegression() #declarando objeto
regressao.fit(x,y) #trieinando objeto
print(f'O coeficiente é : {regressao.coef_} e o interceptador seria: {regressao.intercept_}')
print(f'A equação é: {regressao.coef_}*x +{regressao.intercept_}')
```

```python
y_predi = regressao.predict(x) #fazendo predição dos valores
print(f'Erro Quadrático Médio: = {mean_squared_error(y,y_predi)}')
print(f'Score de Variancia : {r2_score(y, y_predi)}')
```

Outra forma de imprimir os dados com OLS:
(ref.> https://www.statsmodels.org/stable/index.html)

```python
import statsmodels.api as sm
x2 = sm.add_constant(x)
est = sm.OLS(y,x2)
est2 = est.fit()
print(est2.summary())
```

# Regressão Linear Multipla  
Ultilizaremos a mesma biblioteca.

Primeiro vamos redefinir quem sao nossa variaveis preditoras e preditas.

```python
# retira do data frame a minha variavel a ser predita.
x_multipla = dados.drop(['Vendas_Cafe','Promocao'],axis=1)
# ha que remover as variaveis de texto tambem
y = dados['Vendas_Cafe'].values.reshape(-1,1)
```

Inicializamos o objetor de regressão e treinando o modelo.

```python
regressao_mult = linear_model.LinearRegression()
regressao_mult.fit(x_multipla,y)
```

```python
print(f'O coeficiente do {x_multipla.columns.values[0]} é {regressao_mult.coef_.item(0)} e o Coeficiente do {x_multipla.columns.values[1]} é {regressao_mult.coef_.item(1)}')
print(f'O ponto de interceção é: {regressao_mult.intercept_.item(0)}')
print(regressao_mult.score(x_multipla,y))
```

Mostranso os resultados com biblioteca OLS :
(ref.> https://www.statsmodels.org/stable/index.html)

```python
import statsmodels.api as sm
x2 = sm.add_constant(x_multipla) #primeiro recebe os vlaores de X
est = sm.OLS(y, x2)                #aciona o metodo com y e X
est2 = est.fit()                #realiza os calculos
print(est2.summary())           #exibe o relatorio
```

Vamos considerar agora as respostas categoricas quanto as promoções. Se há ou não promoção.

Para isso vamos mudar a coluna promoção para valore de 0 e 1.

0 - Para casos em que não há promoção;

1 - Para casos que há promoção;

```python
#definindo a funcao que vai transformar em binario
def promo_bin(valor:str):
    try:
        if valor in "Nao":
            return 0
        else:
            return 1
    except:
        return valor
```

```python
dados['Promocao'] = dados['Promocao'].apply(promo_bin)
#aqui se aplica a funcao criada a cada item da tabela promo
```

Transformada a coluna em binario, agora vamos definir as variaveis preditoras e preditas.

```python
x_multipla = dados.drop('Vendas_Cafe', axis=1)
y_multipla = dados['Vendas_Cafe'].values.reshape(-1,1)
```

```python
regressao_mult = linear_model.LinearRegression()
regressao_mult.fit(x_multipla, y_multipla)
for i in range(0,len(x_multipla.columns.values)):
    print(f'O coeficiente de é {x_multipla.columns.values[i]} : {regressao_mult.coef_.item(i)}')
```

Mostranso os resultados com biblioteca OLS :
(ref.> https://www.statsmodels.org/stable/index.html)

```python
import statsmodels.api as sm
x2 = sm.add_constant(x_multipla)
est = sm.OLS( y_multipla, x2)
est2 = est.fit()
print(est2.summary())
```
