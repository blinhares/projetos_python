# **Documentar Funçoes em Python**
Realizar a documentaçao correta facilita o entednimento da funcao por terceiros, bem como organiza e facilita o uso posterior da mesma.

A descriçao geral da funcao deve ser colocada entre ''' como mostrado abaixo.

Definindo uma funçao:

````python
def funcao(var1:int , var2:float=21.1, var3:str='valor da variavel'):
  '''Aqui se descreve a funçao e tudo o mais que se faca nescessário para compreende-la.
  Esta informaçao sera mostra ao usuario utilizando o argumento __doc__
  Aqui tambem se poderar descrever detalhadamente as variaveis presentes na funçao
  bem como dar exemplos do seu funcionamento...'''
  print(f'Variavel 01 é : {var1} \n Variavel 02 é: {var2} \n Variavel 03 é: {var3}')
  pass
````

Para visualizar a descriçao da funcao temos que utilizar o metodo doc como mostrado abaixo:

````python
funcao.__doc__
````


Na de claraçao da funçao se pode inserir as variaveis seguida dos tipos e com valores sujeridos em caso de que não sejam informados.
No caso de uma variavel não ter um valor sujerido e náo for fornecido a funcao apresentara erro!

Todas as variaveis e tipos, podem ser consultados atraves do metodo annotations ."

````python
funcao.__annotations__
````

````python
funcao(1)
````

No caso de náo incluir o valor da unica variavel que náo a sujestao de valor, nos deparamos com o seguinte erro:

````python
funcao()
````

# Medindo o Tempo de Execução de Uma Função

Usando biblioteca TIMEIT

Comparando Loopins Vs. Use Comprehensions

````python
import timeit



def do_1():
    list_object = []
    for i in range(100):
        list_object.append(i)

t = timeit.Timer(setup='from __main__ import do_1', stmt='do_1()')
print(f'Tempo para op. 01 : {t.timeit()}')

def do():
    [i for i in range(100)]

t = timeit.Timer(setup='from __main__ import do', stmt='do()')
print(f'Tempo para op. 02 : {t.timeit()}')
````
